```python
# CARGAR BIBLIOTECAS NECESARIAS
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
# CARGAR BIBLIOTECAS NECESARIAS
%matplotlib inline

# Configuración básica
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

# Configuración de gráficos
sns.set_style('whitegrid')
plt.rcParams['figure.figsize'] = (10, 6)

df = pd.read_csv('DatosReforestacionFonafifo.csv', delimiter=";", encoding='ISO-8859-1')
df.head()
```

```python
df.columns
```

```python
grouped_df = df.groupby(['Edad (Meses)', 'Especie (nombre cientifico)'])
mean_g_per_especie = grouped_df['G (M2/ARBOL)'].mean()
unique_especies = df['Especie (nombre cientifico)'].unique()
for especie in unique_especies:
    for edad in grouped_df.groups:
        if especie == edad[1]:
            print(f"Especie: {especie}, Edad: {edad[0]}, Mean G: {mean_g_per_especie[edad]}")
```

```python
df['Especie (nombre cientifico)'] = df['Especie (nombre cientifico)'].replace('Gmelina Arborea', 'Gmelina arborea')
```

```python
df['Especie (nombre cientifico)'] = df['Especie (nombre cientifico)'].replace('Vochysia guatemalensis ', 'Vochysia guatemalensis')
```

```python
suma_g_por_especie = df.groupby('Especie (nombre cientifico)')['G (M2/ARBOL)'].sum().sort_values(ascending=False)
print(suma_g_por_especie)
```

```python
especies = ['Tectona grandis', 'Gmelina arborea', 'Vochysia guatemalensis', 'Pinus caribaea', 'Vochysia guatemalensis', 'Gmelina Arborea', 'Acacia mangium', 'Terminalia amazonia']
observaciones = df[df['Especie (nombre cientifico)'].isin(especies)]['Especie (nombre cientifico)'].value_counts()
print(observaciones)
```

**Resultados Obtenidos**

La base de datos presenta mucha variabilidad y errores. Entre los errores se puede considerar que hay un sistema de referencia en los datos que no fue identificado (ND) cuando lo ideal es que los datos sean georeferenciados en el sistema de proyección CRTM05 o algún sistema de proyección definido para poderlos representar en un mapa. Algunas variables prácticamente nunca han sido medidas ni utilizadas, como por ejemplo factores de forma, estado fitosanitario, tipo de arreglo SAF para los casos de bloques de plantaciones, entre otros. Se puede mejorar y unificar la toma de información para seguir recopilando información más consisa y así generar mejores análisis.

Según el presente análisis, las especies de las que más se ha recopilado información durante los últimos 17 años son Tectona grandis, Gmelina arborea, Vochysia guatemalensis, Pinus caribae, Acacia mangium y Terminalia amazonia. Las dos principales y de las que se ha generado mas información son Tectona grandis y Gmelina arborea, dos especies que prácticamente se plantan en todo el país.

En los gráficos obtenidos por medio de pandas profiling, en general, se observa una fuerte correlación del área basal y la altura, lo cual es muy consecuente con lo que se observa en la realidad y se puede valorar como un indicador significativo del rendimiento en los proyectos de reforestación, sin tener que recurrir a otros factores necesariamente como por ejemplo factores de forma y estado fitosanitario, variables que de cualquier forma pueden volverse un poco subjetivas dependiendo de las referencias del observador.

En un futuro cercano vale la pena analizar, si se puede considerar que las variables de raleo, resiembra y labores de mantenimiento en una plantación forestal realmente generan un mayor beneficio NO en cuanto a crecimiento y rendimiento si no desde el punto de vista de captación de carbono, ya que es una de las variables que se considera muy importante en los últimos años como servicio ambiental.

***Conclusiones***
- Las especies con mayor cantidad y mejor calidad de información obtenida por Fonafifo en los últimos 17 años son Gmelina arborea y Tectona grandis.
- La recopilación de información de caracter cualitativo puede ser mas irregular y subjetiva por lo que no es muy útil para hacer análisis temporal sobre el comportamiento para ninguna especie ni para analizas crecimiento y rendimiento.
- Las variables de área basal (G) y altura total (H) tienen una fuerte correlación por lo que se pueden utilizar para representar el buen rendimiento de una determinada especie en una plantación forestal.
- La mayoría de las observaciones provienen de plataciones no raleadas, por lo que vale la pena considerar que es parte de la realidad y que se puede valorar si el beneficio considerando la captación de carbono es mayor en plantaciones sin raleo que en plantaciones con raleos. Recordando que los raleos son parte del mantenimiento que se le debe dar a las plantaciones forestales como parte del contrato que firman por pago de servicios ambientales (PSA) con FONAFIFO, pero que al final el propietario es quien debe hacer la inversión y obtener el beneficio final del aprovechamiento de la plantación. Mientras que a FONAFIFO le interesa valorar el mayor beneficio ambiental que se obtiene al mantener la plantación en pie, como podría ser el caso de la captación de carbono.
